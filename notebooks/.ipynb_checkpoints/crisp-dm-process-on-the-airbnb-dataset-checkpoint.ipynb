## Entendimiento del Negocio

Airbnb es una plataforma de alojamiento que satisface las necesidades de alojamiento y de préstamo.
Su principal fuente de ingresos es la **comisión del anfitrión**. Básicamente, a medida que aumenta el número de transacciones entre el anfitrión y el huésped, también aumentan sus ganancias.
Por lo tanto, es importante para su negocio y espero que sea uno de sus KPI.

A continuación, extraigo el momento en que se produce el pico cada año.

```python
review_q1_df["year"] = review_q1_df.date.dt.year
years = review_q1_df.year.unique()

for year in years:
    if year >= 2010 and year < 2016:
        year_df = review_q1_df[review_q1_df.year == year]
        max_value = year_df.rolling_mean_30.max()
        max_date = year_df[year_df.rolling_mean_30 == max_value].date.dt.date.values[0]
        print(year, max_date, np.round(max_value, 1))
```

Mi hipótesis es correcta.

## Preparación de datos

A partir de aquí, quiero responder a la última pregunta: "¿Hay alguna tendencia en las habitaciones populares?".

Comencemos a depurar y procesar los datos del anuncio.

Primero, necesitamos definir la variable objetivo "Popularidad de los anuncios". Pensé que podría definirse de la siguiente manera:

[Número real de alquileres] / ([Días disponibles de 2016 a 2017] * (2017 - [Año de apertura de los anuncios]))

Dado que el número de alquileres de los anuncios se considera proporcional al número de días que la propiedad está disponible, es necesario escalarlo. Además, dado que se considera que el número aumenta cuanto antes se publiquen los anuncios, también debemos escalarlo.

Un aspecto que no puedo considerar aquí es el período en que los anuncios estuvieron realmente alquilados. Cuanto más largo sea el período de alquiler, menor será el número de alquileres. Sin embargo, no hay datos que lo consideren, por lo que asumiré que la mayoría de los usuarios de Airbnb son usuarios de corta estancia.

Primero, verificamos si hay valores nulos. Esto se debe a que no podemos usar columnas con demasiados valores nulos.

```python
prepare_df = seattle_listing.copy(deep=True)
```

```python
# check null count
df_length = prepare_df.shape[0]

for col in prepare_df.columns:
    null_count = prepare_df[col].isnull().sum()
    if null_count == 0:
        continue
        
    null_ratio = np.round(null_count/df_length * 100, 2)
    print("{} has {} null values ({}%)".format(col, null_count, null_ratio))
```

Como puede ver, parece que hay columnas con un 90 % o más de valores nulos, pero la mayoría tiene entre un 0 % y un 30 % de null_ratio. Por lo tanto, decidí excluir del análisis las columnas con un 30 % o más de null_ratio.

Además, parece que solo hay dos valores nulos en `host_since`, utilizados para calcular la variable objetivo. Esto debe eliminarse.

```python
# detect need drop columns
drop_cols = [col for col in prepare_df.columns if prepare_df[col].isnull().sum()/df_length >= 0.3]

# drop null
prepare_df.drop(drop_cols, axis=1, inplace=True)
prepare_df.dropna(subset=['host_since'], inplace=True)

# check after
for col in prepare_df.columns:
    null_count = prepare_df[col].isnull().sum()
    if null_count == 0:
        continue
        
    null_ratio = np.round(null_count/df_length * 100, 2)
    print("{} has {} null values ({}%)".format(col, null_count, null_ratio))
```

Bien.
A continuación, eliminaremos la columna que no parece ser una característica de facilidad de préstamo. Dado que esta vez no incluimos el procesamiento del lenguaje natural, también eliminaremos las columnas basadas en comentarios.

```python
drop_cols = ['listing_url', 'scrape_id', 'last_scraped', 'name', 'summary', 'space', 'description', 'neighborhood_overview',
                'transit', 'medium_url', 'picture_url', 'xl_picture_url', 'host_id', 'host_url', 'host_name', 'host_about', 'host_thumbnail_url',
                'host_picture_url', 'street', 'city', 'state', 'zipcode', 'market', 'smart_location', 'country_code', 'country', 'latitude', 'longitude',
                'calendar_updated', 'calendar_last_scraped', 'first_review', 'last_review', 'amenities', 'host_verifications']

prepare_df.drop(drop_cols, axis=1, inplace=True)
```

```python
prepare_df.columns
```

```python
drop_cols = []
for col in prepare_df.columns:
    if prepare_df[col].nunique() == 1:
        drop_cols.append(col)
        
prepare_df.drop(drop_cols, axis=1, inplace=True)
prepare_df.columns
```

Bien. Ya casi termino de eliminar las columnas que no se usan para el análisis. A continuación, creo un objetivo valioso.


```python
print('number_of_reviews' in prepare_df.columns)
print(prepare_df[['id', 'number_of_reviews']].head())
```

```python
# Calcular días disponibles
listing_available = seattle_calendar.groupby('listing_id')['price'].count().reset_index()
listing_available.columns = ["id", "available_count"]

# Merge con prepare_df
prepare_df = prepare_df.merge(listing_available, how='left', on='id')

# Calcular tasa de ocupación
prepare_df['host_since_year'] = pd.to_datetime(prepare_df['host_since']).dt.year
prepare_df['occupancy_rate'] = prepare_df['number_of_reviews'] / (prepare_df['available_count'] + 1)

```

A continuación, elimine la columna que parezca estar directamente relacionada con la variable objetivo (por ejemplo, num_review).

```python
print("Before: {} columns".format(prepare_df.shape[1]))

drop_cols = ['host_since', 'accommodates', 'availability_30', 'availability_60', 'availability_90', 'availability_365',
                'number_of_reviews', 'review_scores_rating', 'available_count', 'reviews_per_month', 'host_since_year', 'review_scores_value']

prepare_df.drop(drop_cols, axis=1, inplace=True)
print("After: {} columns".format(prepare_df.shape[1]))
```

A partir de aquí, convertiré los datos a un formato que el modelo pueda aprender. Primero, convertiré la categoría "valores" en valores ficticios.

```python
# convert true or false value to 1 or 0
dummy_cols = ['host_is_superhost', 'require_guest_phone_verification', 'require_guest_profile_picture', 'instant_bookable', 
              'host_has_profile_pic', 'host_identity_verified', 'is_location_exact']

for col in dummy_cols:
    prepare_df[col] = prepare_df[col].map(lambda x: 1 if x == 't' else 0)

# create dummy valuables
dummy_cols = ['host_location', 'host_neighbourhood', 'neighbourhood', 'neighbourhood_cleansed', 'neighbourhood_group_cleansed',
             'property_type', 'room_type', 'bed_type', 'cancellation_policy', 'host_response_time']

prepare_df = pd.get_dummies(prepare_df, columns=dummy_cols, dummy_na=True)
```

```python
df_length = prepare_df.shape[0]

for col in prepare_df.columns:
    null_count = prepare_df[col].isnull().sum()
    if null_count == 0:
        continue
        
    null_ratio = np.round(null_count/df_length * 100, 2)
    print("{} has {} null values ({}%)".format(col, null_count, null_ratio))
```

A continuación, gestione los valores nulos restantes.

```python
prepare_df["is_thumbnail_setted"] = 1 - prepare_df.thumbnail_url.isnull()
prepare_df.drop('thumbnail_url', axis=1, inplace=True)
prepare_df.host_response_rate = prepare_df.host_response_rate.fillna('0%').map(lambda x: float(x[:-1]))
prepare_df.host_acceptance_rate = prepare_df.host_acceptance_rate.fillna('0%').map(lambda x: float(x[:-1]))
prepare_df.bathrooms.fillna(0, inplace=True)
prepare_df.bedrooms.fillna(0, inplace=True)
prepare_df.beds.fillna(0, inplace=True)
prepare_df.cleaning_fee.fillna('$0', inplace=True)
prepare_df.review_scores_accuracy.fillna(0, inplace=True)
prepare_df.review_scores_cleanliness.fillna(0, inplace=True)
prepare_df.review_scores_checkin.fillna(0, inplace=True)
prepare_df.review_scores_communication.fillna(0, inplace=True)
prepare_df.review_scores_location.fillna(0, inplace=True)
```

Finalmente, dado que el valor que se va a representar como un valor numérico se reconoce como una cadena, vamos a convertirlo.

```python
for col in prepare_df.columns:
    if prepare_df[col].dtypes == 'object':
        print(col)
```

```python
prepare_df.price = prepare_df.price.map(lambda x: float(x[1:].replace(',', '')))
prepare_df.cleaning_fee = prepare_df.cleaning_fee.map(lambda x: float(x[1:].replace(',', '')))
prepare_df.extra_people = prepare_df.extra_people.map(lambda x: float(x[1:].replace(',', '')))
```

## Modelado, Evaluación

Este problema es de regresión. Usaré un regresor de bosque aleatorio y usaré validación cruzada para la evaluación.

```python
X = prepare_df.drop(['id', 'easily_accomodated'], axis=1)
y = prepare_df.easily_accomodated.values

rf = RandomForestRegressor(n_estimators=100, max_depth=5)
scores = cross_val_score(rf, X, y, cv=5)
```

```python
scores
```

No puedo predecir nada. ¿Por qué? Para entender por qué, hagamos un diagrama de dispersión con los valores predichos y reales.

```python
rf.fit(X, y)
predictions = rf.predict(X)
```

```python
plt.figure(figsize=(8, 8))

plt.plot((0, 4), (0, 4), color='gray')
plt.plot(y, predictions, linewidth=0, marker='o', alpha=0.5)
plt.grid()
plt.xlim((-0.2, 4.2))
plt.ylim((-0.2, 4.2))
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.show()
```

Cuanto más cerca de la línea gris, más preciso es el valor predicho.
Entonces, a primera vista, puedo entender que los valores predichos no son precisos.
Además, también se entiende que la predicción se ajusta a un valor pequeño cercano a 0, y se realiza una predicción pequeña para el valor mayor.

Intentemos la transformación logarítmica de la variable objetivo.

```python
X = prepare_df.drop(['id', 'easily_accomodated'], axis=1)
y = np.log(prepare_df.easily_accomodated.values)

rf = RandomForestRegressor(n_estimators=100, max_depth=5)
scores = cross_val_score(rf, X, y, cv=5)
print(scores)
```

```python
rf.fit(X, y)
predictions = rf.predict(X)
```

```python
plt.figure(figsize=(8, 8))

plt.plot((-10, 10), (-10, 10), color='gray')
plt.plot(y, predictions, linewidth=0, marker='o', alpha=0.5)
plt.grid()
plt.xlim((-8.2, 2))
plt.ylim((-8.2, 2))
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.show()
```

Ha mejorado un poco, pero los valores siguen siendo demasiado densos y se han incluido muchos valores predichos en esa parte.
Para mejorar esto, creo que dos métodos son eficaces.

* Revisar la lógica de la variable objetivo; esta definición puede ser demasiado compleja; sería mejor con una definición más simple de los valores objetivo.
* Realizar un submuestreo en zonas densas.
